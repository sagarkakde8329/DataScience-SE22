```python
from collections import deque

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [], 'E': [], 'F': []
}

def bfs(start):
    visited, queue = set(), deque([start])
    print("BFS:", end=" ")
    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            queue.extend(graph[node])

def dfs(start, visited=set()):
    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for n in graph[start]:
            dfs(n, visited)

bfs('A')
print("\nDFS:", end=" ")
dfs('A')
```

```python
import heapq

def a_star(start, goal, graph, h):
    pq = [(0 + h[start], 0, start)]
    visited = set()

    while pq:
        f, g, node = heapq.heappop(pq)
        if node == goal:
            print("Reached:", node)
            return
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            for n, cost in graph[node]:
                heapq.heappush(pq, (g + cost + h[n], g + cost, n))

graph = {
    'A': [('B',1), ('C',3)],
    'B': [('D',1)],
    'C': [('D',1)],
    'D': []
}

heuristic = {'A':3, 'B':2, 'C':1, 'D':0}

print("\nA* Traversal:")
a_star('A', 'D', graph, heuristic)
```

```python
def minimax(depth, node, is_max, scores, max_depth):
    if depth == max_depth:
        return scores[node]

    if is_max:
        return max(minimax(depth+1, node*2, False, scores, max_depth),
                   minimax(depth+1, node*2+1, False, scores, max_depth))
    else:
        return min(minimax(depth+1, node*2, True, scores, max_depth),
                   minimax(depth+1, node*2+1, True, scores, max_depth))

scores = [3, 5, 2, 9, 12, 5, 23, 23]
depth = 3

print("\nMinimax Optimal Depth:", minimax(0, 0, True, scores, depth))
```
