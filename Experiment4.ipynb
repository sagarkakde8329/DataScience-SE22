```python
import numpy as np
import matplotlib.pyplot as plt


data_normal = np.random.normal(loc=0, scale=1, size=10000)

plt.hist(data_normal, bins=50, edgecolor='black', alpha=0.7)

plt.title('Histogram of Normal Distribution')

plt.xlabel('Value')
plt.ylabel('Frequency')

plt.grid(True)
plt.show()

print("Normal distribution dataset generated and histogram plotted successfully.")
```

## Compute Descriptive Statistics

### Subtask:
Create a sample dataset and then compute its mean, median, variance, and standard deviation using `numpy`.


```python
import numpy as np

sample_data = np.array([10, 12, 23, 23, 16, 23, 21, 16, 29, 21, 22, 19])

mean_value = np.mean(sample_data)

median_value = np.median(sample_data)

variance_value = np.var(sample_data)

std_dev_value = np.std(sample_data)

print(f"Sample Data: {sample_data}")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Variance: {variance_value}")
print(f"Standard Deviation: {std_dev_value}")

print("Descriptive statistics computed successfully.")
```

## Simulate Coin Toss Experiment

### Subtask:
Simulate a series of coin tosses (e.g., 1000 tosses) and compute the probabilities of heads and tails. Display the results.


```python
import random

num_tosses = 10000

heads_count = 0
tails_count = 0

for _ in range(num_tosses):
    outcome = random.randint(0, 1)
    if outcome == 1:
        heads_count += 1
    else:
        tails_count += 1

probability_heads = heads_count / num_tosses

probability_tails = tails_count / num_tosses

print(f"Total number of tosses: {num_tosses}")
print(f"Number of heads: {heads_count}")
print(f"Number of tails: {tails_count}")
print(f"Probability of heads: {probability_heads:.4f}")
print(f"Probability of tails: {probability_tails:.4f}")

print("Coin toss experiment simulated and probabilities computed successfully.")
```
