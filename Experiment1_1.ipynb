### Factorial of a Number

This code calculates the factorial of a given non-negative integer using a recursive function.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
num = 5
print(f"The factorial of {num} is {factorial(num)}")

num = 0
print(f"The factorial of {num} is {factorial(num)}")
```

### Expert System for Weather-Based Activity Suggestions

This code defines a simple rule-based expert system that suggests activities based on the current weather and the day of the week.

```python
def suggest_activity(weather, day):
    weather = weather.lower()
    day = day.lower()

    if weather == 'sunny':
        if day == 'weekend':
            return "It's sunny and the weekend! How about a picnic in the park or a hike?"
        else:
            return "Sunny day! Consider a walk during your break or after work."
    elif weather == 'rainy':
        if day == 'weekend':
            return "Rainy weekend. Perfect for reading a book, watching movies, or visiting a museum."
        else:
            return "Rainy day. Stay indoors, perhaps work on a creative project or enjoy a hot beverage."
    elif weather == 'cloudy':
        if day == 'weekend':
            return "Cloudy weekend. Maybe a leisurely stroll, visiting a local cafe, or a board game session."
        else:
            return "Cloudy day. A good time for indoor tasks or a quiet evening at home."
    elif weather == 'snowy':
        if day == 'weekend':
            return "Snowy weekend! Time for building a snowman, sledding, or cozying up by the fire."
        else:
            return "Snowy day. Be careful outside! Perhaps enjoy the view from indoors with a warm drink."
    else:
        return "Weather condition not recognized. Please try 'sunny', 'rainy', 'cloudy', or 'snowy'."

# Example Usage:
print(f"Today: Sunny, Weekend - Suggestion: {suggest_activity('Sunny', 'Weekend')}")
print(f"Today: Rainy, Weekend - Suggestion: {suggest_activity('Rainy', 'Weekend')}")
print(f"Today: Cloudy, Weekday - Suggestion: {suggest_activity('Cloudy', 'Weekday')}")
print(f"Today: Snowy, Weekend - Suggestion: {suggest_activity('Snowy', 'Weekend')}")
print(f"Today: Sunny, Weekday - Suggestion: {suggest_activity('Sunny', 'Weekday')}")
print(f"Today: Windy, Weekday - Suggestion: {suggest_activity('Windy', 'Weekday')}")
```

### Palindrome Check

This code defines a function to check if a given string is a palindrome (reads the same forwards and backward). It ignores case and non-alphanumeric characters for a more robust check.

```python
import re

def is_palindrome(text):
    # Convert to lowercase and remove non-alphanumeric characters
    processed_text = re.sub(r'[^a-zA-Z0-9]', '', text).lower()
    return processed_text == processed_text[::-1]

# Example usage:
print(f"'madam' is a palindrome: {is_palindrome('madam')}")
print(f"'A man, a plan, a canal: Panama!' is a palindrome: {is_palindrome('A man, a plan, a canal: Panama!')}")
print(f"'hello' is a palindrome: {is_palindrome('hello')}")
print(f"'Racecar' is a palindrome: {is_palindrome('Racecar')}")
print(f"'12321' is a palindrome: {is_palindrome('12321')}")
```

### Odd or Even Number Check

This code defines a function to check if a given number is odd or even and then demonstrates its use with examples.

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
print(check_odd_even(4))
print(check_odd_even(7))
print(check_odd_even(0))
print(check_odd_even(-3))
```
