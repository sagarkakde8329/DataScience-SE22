# Data Manipulation using Open-Source Dataset in Python: Iris Dataset Experiment

## Objective
This experiment aims to demonstrate fundamental data manipulation techniques using the Python libraries `pandas` and `numpy` on the classic Iris dataset. We will cover loading data, inspecting its structure, performing basic statistical analysis, filtering records, and aggregating data.

```python

import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

print("Libraries imported successfully.")
```

```python
iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

df['species'] = iris.target

species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species'] = df['species'].map(species_map)

print("Iris dataset loaded into a pandas DataFrame.")

print("\nFirst 5 rows of the dataset:")
display(df.head())
```

```python
print("\nDataset Information:")
df.info()

print("\nColumn Names:")
print(df.columns.tolist())

print(f"\nNumber of records (rows): {df.shape[0]}")
print(f"Number of features (columns): {df.shape[1]}")
```

```python
print("\nBasic Statistical Summary of the dataset:")
display(df.describe())
```

```python
filtered_df = df[df['petal length (cm)'] > 5]

print("\nDataset filtered for 'petal length (cm)' > 5:")
display(filtered_df.head())
print(f"Number of records after filtering: {filtered_df.shape[0]}")
```

```python
aggregated_stats = df.groupby('species').agg(
    mean_sepal_length=('sepal length (cm)', 'mean'),
    min_sepal_width=('sepal width (cm)', 'min'),
    max_petal_length=('petal length (cm)', 'max'),
    count_samples=('sepal length (cm)', 'count')
)

print("\nAggregate statistics (mean, min, max, count) grouped by species:")
display(aggregated_stats)
```

## Result

The experiment successfully demonstrated various data manipulation techniques using the Iris dataset.

1.  **Dataset Loading and Inspection**: The Iris dataset was loaded into a pandas DataFrame, and its first 5 rows were displayed, showing the `sepal length`, `sepal width`, `petal length`, `petal width`, and `species` columns.
2.  **Data Information**: `df.info()` provided a concise summary, including column data types (float64 for features, object for species) and non-null counts, confirming no missing values. The `df.columns` output listed all column names, and `df.shape` confirmed 150 records and 5 features.
3.  **Statistical Summary**: `df.describe()` generated descriptive statistics for numerical columns, showing counts, means, standard deviations, min/max values, and quartiles, giving an initial understanding of data distribution.
4.  **Filtering**: A new DataFrame `filtered_df` was created by selecting rows where 'petal length (cm)' was greater than 5. This demonstrated how to subset data based on a condition, reducing the dataset from 150 to 49 records.
5.  **Grouping and Aggregation**: The dataset was successfully grouped by the 'species' column. For each species, aggregate statistics (mean sepal length, minimum sepal width, maximum petal length, and count of samples) were calculated, revealing distinct characteristics for 'setosa', 'versicolor', and 'virginica' species. For instance, 'setosa' has significantly smaller petal lengths compared to 'versicolor' and 'virginica'.
